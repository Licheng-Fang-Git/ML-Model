
# **Exploratory Data Analysis**

        - Method used to analyze data sets and summarize their main characteristics

        - Transforming and cleaning data to make it more useful

        - Makeing it easier to discover patterens, spot anomalies, checking assumptions or even testing hypothesis
    
# **Two Subgroups of EDA**

        - Uni-Variat
    
        - Multi-Variat

# **Uni-Variat Analysis**

        - Purpose: Describe the data and find patterns that exist within it. 

        - Non Graphical:
        
                - Since it deals with a single variable, its purpose is to describe the data's central tendency (mean, median) and spread (variance, standard deviation) without exploring relationships or causality

        - Graphical:

                - Common Types of graphs include:
        
                        - Stem and Leaf Plots which shows all the data values and the shape of distribution

                        - Histograms/Bar Plot which each bar represents the porportion of cases for a range of values

                        - Box plot is a fantastic graphical tool for univariate analysis as it clearly shows the median, quartiles, and outliers.
                        


# **Multi-Variat Analysis**

        - Non Graphical:

                - Shows relationship between two or more varaibles of data through cross tabulation or statistics 
        
        - Multivariate Graphics:
        
                - Group bar charts which each group represents one level of the other variable
                
                - Bubble charts, heat maps and run charts

                - Bubble Charts and Heat Maps: Useful for visualizing relationships among three or more variables.

# **Step 3: Feature Understanding**

```python
# Convert the Time Intervals to hour only
# def time_to_hours(t):
#     return pd.to_datetime(t, format="%I:%M %p").hour + pd.to_datetime(t, format="%I:%M %p").minute / 60

# df['Interval Times'] = df['Interval Times'].apply(time_to_hours)

plt.figure(figsize=(15, 6))
sns.regplot(data=df, x='Interval Times', y='Impressions', scatter_kws={'alpha': 0.5, 'color': 'red'}, line_kws={'color': 'blue'})
plt.xticks(range(0, 24), [f"{h % 12 or 12} {'AM' if h < 12 else 'PM'}" for h in range(0, 24)])
plt.grid(True)
plt.show()
```

1. Overall Relationship
The blue regression line shows a slight positive slope — meaning posts made later in the day tend to get a bit more impressions on average.

However, the confidence interval (blue shaded area) is wide in the early hours (e.g., before 9 AM), suggesting there’s high uncertainty due to fewer data points.

2. Posting Patterns
There’s a clear cluster between 10 AM – 3 PM, indicating that most posts happen during late morning to mid-afternoon.

Very few posts occur before 8 AM or after 6 PM, so any trend there is not statistically reliable.

3. High-Impression Outliers
There are several outlier posts with impressions above 15,000–20,000, especially around 10 AM, 11 AM, 1 PM, and 2 PM.

These high-performing posts might be skewing the regression line upward — worth checking if they share similar content type or category.

4. Practical Insights
Midday seems to be the sweet spot for posting (both frequency and some top-performers are there).

Early morning and evening data are too sparse to draw strong conclusions, so we’d need more posts in those slots to confirm if they’re actually underperforming.

```python
# Scatterplots with regression lines
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
x = df['Likes']
y = df['Impressions']
slope, intercept, r_value, p_value, std_err = linregress(x, y)
sns.regplot(data=df, x='Likes', y='Impressions', ax=axes[0], scatter_kws={'alpha':0.5})
axes[0].text(x.min(), y.max(), f'$R^2$ = {r_value**2:.3f}', fontsize=12, color='blue')

axes[0].set_title('Likes vs Impressions')

x = df['Comments']
y = df['Impressions']
slope, intercept, r_value, p_value, std_err = linregress(x, y)
sns.regplot(data=df, x='Comments', y='Impressions', ax=axes[1], scatter_kws={'alpha':0.5})
axes[1].text(x.min(), y.max(), f'$R^2$ = {r_value**2:.3f}', fontsize=12, color='blue')
axes[1].set_title('Comments vs Impressions')

x = df['Clicks']
y = df['Impressions']
slope, intercept, r_value, p_value, std_err = linregress(x, y)
sns.regplot(data=df, x='Clicks', y='Impressions', ax=axes[2], scatter_kws={'alpha':0.5})
axes[2].text(x.min(), y.max(), f'$R^2$ = {r_value**2:.3f}', fontsize=12, color='blue')
axes[2].set_title('Clicks vs Impressions')

plt.tight_layout()
plt.show()
```

Likes vs. Impressions:

Strong positive linear trend: As likes increase, impressions rise sharply.

The points are relatively tight around the regression line, indicating a strong correlation.

Suggests that likes are a solid proxy for reach — either driving impressions or co-occurring due to high-quality content.

Takeaway: Likes are a strong indicator or driver of impressions.

Comments vs. Impressions

Still a positive trend, but noticeably weaker:

    - Fewer comments overall (most points clustered below 5).

    - The regression line is influenced by a few outliers (high-comment, high-impression posts).

The wider confidence band shows higher variability and lower predictive power than likes.

Takeaway: Comments may boost impressions, but are less consistent and more sparse.

Clicks vs. Impressions

A strong positive correlation, similar in shape to the Likes plot.

Some high click count outliers (e.g., 3000–5000 clicks) with correspondingly high impressions — these may skew the slope.

The confidence band is tighter in the lower range (common values), suggesting reliability there.

Takeaway: Clicks are also strongly associated with impressions, especially for high-performing posts.

```python
# Boxplot of Impressions by Post Type

df2 = df.copy()
df2 = df2.dropna(subset=['Type of Post'])

plt.figure(figsize=(10,5))
sns.boxplot(data=df2, x='Type of Post', y='Impressions', hue='Type of Post', palette='Set2')
plt.title('Impressions by Post Type')
plt.show()
list_post = list(map(str,df2['Type of Post'].unique()))
for i in range(len(list_post)):
    impression_data = df2[df2['Type of Post'] == list_post[i]]['Impressions']
    min_value = impression_data.min()
    max_value = impression_data.max()
    range_value = max_value - min_value
    Q1 = impression_data.quantile(0.25)
    Q3 = impression_data.quantile(0.75)
    IQR = Q3 - Q1
    print(f"Statistics for {list_post[i]}:")
    print(f"  Min: {min_value}")
    print(f"  Max: {max_value}")    
    print(f"  Range: {range_value:}")
    print(f"  Q1: {Q1}")
    print(f"  Q3: {Q3}")
    print(f"  IQR: {IQR}")
data = {
    "Post Type": ['Video', 'Text', 'Image', 'Repost'],
    'Impressions Median': ['High', 'Moderate-High', 'Moderate', 'Very Low'],
    'Range & Outliers': ['Wide + strong outliers', 'Tight; consistent', 'Very high outliers', 'None'],
    "Performance Verdict": ['Best overall', 'Reliable', 'Volatile', 'Avoid']
}
display(pd.DataFrame(data))
```

Video

    IQR: 4294 → widest spread, indicating unpredictable performance.

    Very high max (19,962) shows viral potential, but it’s not consistent.

    Suitable for high-risk, high-reward strategy.

Image

    IQR: 3464.5, second highest.

    Also has very high max (20,272), meaning images can perform very well.

    Slightly more consistent than video, but still volatile.

Text
    IQR: 1822.75 → most consistent among high-performers.

    Performance is tight and reliable, although not as explosive as video/image.

    Best for steady reach without high variance.

Repost

    Extremely low IQR (125) and range.

    Consistently low-performing, no outliers.

    Likely not worth prioritizing unless strategic (e.g., internal resharing).


Trendlines show that Tuesday and Thursday posts at 12–2 PM get better impressions.

Actionable Insights

Target Posting Time:
Aim for 12:00 to 14:00 (midday) — it’s the most active and statistically stable period.

Avoid Late Fridays:
Impressions trend down as Friday progresses — earlier posts perform better.

Test Early Mondays:
Try a few early morning Monday posts (9–10 AM) to see if you can capture underused time windows.

Use Tuesdays Strategically:
Steady positive slope implies later posts (e.g., 1–2 PM) perform well on Tuesdays.
